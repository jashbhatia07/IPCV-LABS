# **Read Image**

```python
import cv2 
import matplotlib.pyplot as plt
image = cv2.imread('sea_link.jpg') 
image
```

```python
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

# **Display Image**

```python
plt.imshow(img)
```

# **Gray-Scale image**

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
gray_image.shape
```

# **CROP**

```python
crop = gray_image[50:180, 100:300]  
plt.imshow(crop)
```

# **Arithmetic Operations**

```python
image1 = cv2.imread('gateway.jpg') 
image2 = cv2.imread('sea_link.jpg')
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
print(image1.shape)
print(image2.shape)
```

```python
plt.imshow(image1)
```

```python
plt.imshow(image2)
```

```python
image2 = cv2.resize(image2, (612, 408))
image2.shape
```

```python
weightedSum = cv2.addWeighted(image1, 0.5, image2, 0.4, 0)
plt.imshow(weightedSum)
```

```python
sub = cv2.subtract(image1, image2)
plt.imshow(sub)
```

# **Logical Operations**

```python
# Bitwise AND
bitwise_and = cv2.bitwise_and(image2, image1)
plt.imshow(bitwise_and)
```

```python
# Bitwise OR
bitwise_or = cv2.bitwise_or(image2, image1)
plt.imshow(bitwise_or)
```

```python
# Bitwise XOR
bitwise_xor = cv2.bitwise_xor(image2, image1)
plt.imshow(bitwise_xor)
```

```python
# Bitwise NOT
bitwise_not = cv2.bitwise_not(image2)
plt.imshow(bitwise_not)
```
