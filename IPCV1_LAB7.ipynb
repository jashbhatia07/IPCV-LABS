**1. Calculating Frequency**

```python
import numpy as np
(unique, counts) = np.unique(img, return_counts=True)
frequencies = np.asarray((unique, counts)).T 
final_frequency = []
final_frequency_key = [] 
for i in frequencies: 
  final_frequency_key.append(i[0])
  final_frequency.append(i[1])
```

```python
for i in range(0, 256): 
  if i not in final_frequency_key: 
    final_frequency_key.insert(i, i) 
    final_frequency.insert(i, 0)
```

**2. Calculating PDF**

```python
for i, j in zip(rang, final_frequency): 
  pdf.append(round(j/sum(final_frequency), 3)) 
print(pdf)
```

**3. Calculating CDF**

```python
cdf = [pdf[0]] 
for i in range(1, len(pdf)): 
  cdf.append(pdf[i] + cdf[-1])
```

```python
cdf[-1]
```

**4. Calculating New Frequency Value**

```python
l_minus_1 = [] 
for i in cdf: 
  i = i * 255 
  l_minus_1.append(int(round(i, 0)))
print(l_minus_1)
```
