```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('6.jpg', 0) 
plt.imshow(img, cmap='gray')
```

```python
img.shape
```

```python
filter = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])
filter
```

```python
output = []
for r in range(img.shape[0]-2): 
  temp = []
  for c in range(img.shape[1]-2):
    lhs = img[r:r+3, c:c+3]
    ans = 0 
    for r_f in range(filter.shape[0]): 
      for c_f in range(filter.shape[1]):
        ans += lhs[r_f][c_f]*filter[r_f][c_f] 
    temp.append(ans) 
  output.append(temp)
```

```python
fig=plt.figure(dpi=200)

fig.add_subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,2,2)
plt.imshow(output,cmap='gray')
plt.axis("off")
plt.title("High Pass Filter")
```

```python
filter_hb = np.array([[-1, -1, -1], [-1, 8.9, -1], [-1, -1, -1]])
filter_hb
```

```python
output_hb = []
for r in range(img.shape[0]-2): 
  temp = []
  for c in range(img.shape[1]-2):
    lhs = img[r:r+3, c:c+3]
    ans = 0 
    for r_f in range(filter_hb.shape[0]): 
      for c_f in range(filter_hb.shape[1]):
        ans += lhs[r_f][c_f]*filter_hb[r_f][c_f] 
    temp.append(ans) 
  output_hb.append(temp)
```

```python
fig=plt.figure(dpi=200)

fig.add_subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.axis("off")
plt.title("Original")

fig.add_subplot(1,2,2)
plt.imshow(output_hb,cmap='gray')
plt.axis("off")
plt.title("High Boost Filter")
```
